# 1.3: Introduction to Neural Networks (NN)

A **Neural Network (NN)** is a machine learning model inspired by the structure and function of the **human brain**.
It is designed to recognize patterns, make decisions, and solve complex problems by learning from data.

### Structure of a Neural Network
**A neural network consists of three main types of layers:**

* Input Layer:

   - Receives the raw data for the model.
   - Each neuron in this layer corresponds to one feature of the input data.
   - Example: In image recognition, each pixel value is an input neuron.
- Hidden Layers:

   - These layers process the data by transforming it into higher-level features.
   - Each hidden layer applies a combination of weights, biases, and activation functions.
   - Example: Detects shapes, textures, or patterns in an image.
- Output Layer:

   - Produces the final prediction or classification.
   - The number of neurons here corresponds to the possible outcomes or labels.
   - Example: For binary classification, the output layer might have 1 neuron representing a probability.

### How Neural Networks Learn:
*Neural networks learn by iteratively improving their predictions through a process called **training**.*

Here's how it works:


**1. Forward Propagation:**

- Input data flows through the network, layer by layer, to produce an output.
- Each neuron applies a weighted sum of inputs, adds a bias, and passes the result through an activation function.

# Activation Functions
**Activation Functions** introduce non-linearity into neural networks, allowing them to learn and represent complex patterns. Each neuron processes its weighted input through an activation function to produce an output.

## Sigmoid Function

$$
œÉ(z) = \frac{1}{1 + e^{-z}}
$$


- Range: (0, 1)
- Use Case: Binary classification, probability outputs.

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
  """Sigmoid function.

  Args:
    x: Input value.

  Returns:
    The sigmoid value.
  """
  return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the sigmoid function
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sigmoid Function')
plt.axhline(y=0.5, color='r', linestyle='--', label='Threshold')
plt.axvline(x=0, color='g', linestyle='--', label='Zero Crossing')
plt.xlabel('x')
plt.ylabel('œÉ(x)')
plt.title('Sigmoid Function')
plt.legend()
plt.grid(True)
plt.show()
```

##### Pros:

- Smooth curve.
- Outputs are easy to interpret as probabilities.

## Tanh (Hyperbolic Tangent)

**Formula:**
$$tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$$

**Range:** (-1, 1)

**Use Case:** Hidden layers, zero-centered data.

**Pros:**
* Outputs are zero-centered, improving optimization.

**Cons:**
* Suffers from vanishing gradient for large or small values of z.

**Example:**
* Input: z = 1
* Output: tanh(1) ‚âà 0.761

```python
import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.linspace(-5, 5, 100)
y = tanh(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Tanh Function')
plt.grid(True)
plt.show()
```

## ReLU (Rectified Linear Unit)

**Formula:**
  
$$
f(z) =
\begin{cases}
z, & \text{if } z > 0 \\\\
0, & \text{if } z \leq 0
\end{cases}
$$

**Range:** [0, ‚àû)

**Use Case:** Hidden layers in deep learning.

**Pros:**
* Avoids vanishing gradient for positive values.
* Computationally efficient.

**Cons:**
* Dying ReLU Problem: Neurons can "die" if they output 0 constantly (negative inputs).

**Example:**
* Input: z = -3, Output: f(z) = 0
* Input: z = 2, Output: f(z) = 2

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
  return np.maximum(0, x)

x = np.linspace(-5, 5, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.grid(True)
plt.show()
```

## Softmax Function

**Formula:**
$$Softmax(z_i) = \frac{e^{z_i}}{\sum_{j=1}^n e^{z_j}}$$

**Range:** (0, 1)

**Use Case:** Multi-class classification.

**How It Works:**
* Converts logits into probabilities for each class.
* The class with the highest probability is chosen.

**Example:**
* Input logits: z = [2, 1, 0]
* Output:
  * Softmax(2) ‚âà 0.79
  * Softmax(1) ‚âà 0.18
  * Softmax(0) ‚âà 0.03

The output probabilities for all classes sum to 1.


# Key Comparisons


| Function | Range | Use Case | Pros | Cons |
|---|---|---|---|---|
| Sigmoid | (0, 1) | Binary classification | Probabilistic output | Vanishing gradient |
| ReLU | [0, ‚àû) | Hidden layers | Efficient, non-linear | Dying neurons (0 outputs) |
| Tanh | (-1, 1) | Zero-centered data | Zero-centered | Vanishing gradient |
| Softmax | (0, 1) | Multi-class classification | Normalized probabilities | Computational cost |

# 1. Perceptron Model Implementation

```python
import tensorflow as tf
import numpy as np

# Input data (e.g., a simple OR gate)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input features
y = np.array([[0], [1], [1], [1]])              # Target labels

# Build the Perceptron Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_dim=2, activation=lambda x: tf.where(x >= 0, 1.0, 0.0))  # Step function
])

# Compile the model (loss and optimizer aren't used for step functions, but are mandatory)
model.compile(optimizer='adam', loss='mse')

# Print the model summary
model.summary()

# Predict (Training isn't needed for a step function here)
predictions = model.predict(X)
print("Predictions:\n", predictions)
```

# 2. Activation Functions with TensorFlow


```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate data (as a tensor)
x = tf.linspace(-10.0, 10.0, 100)  # Use tf.linspace for tensors

# Compute activations
sigmoid = tf.keras.activations.sigmoid(x)
relu = tf.keras.activations.relu(x)
tanh = tf.keras.activations.tanh(x)
softmax = tf.keras.activations.softmax(tf.expand_dims(x, axis=0))  # Softmax requires 2D tensor

# Convert tensors to NumPy arrays for plotting (optional)
sigmoid_np = sigmoid.numpy()
relu_np = relu.numpy()
tanh_np = tanh.numpy()
softmax_np = softmax.numpy().flatten()  # Flatten for 1D plotting

# Plot the activations
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x.numpy(), sigmoid_np, label="Sigmoid")  # Use numpy() for plotting
plt.title("Sigmoid Activation")
plt.grid()
plt.legend()

# ReLU
plt.subplot(2, 2, 2)
plt.plot(x.numpy(), relu_np, label="ReLU")  # Use numpy() for plotting
plt.title("ReLU Activation")
plt.grid()
plt.legend()

# Tanh
plt.subplot(2, 2, 3)
plt.plot(x.numpy(), tanh_np, label="Tanh")  # Use numpy() for plotting
plt.title("Tanh Activation")
plt.grid()
plt.legend()

# Softmax
plt.subplot(2, 2, 4)
plt.plot(x.numpy(), softmax_np, label="Softmax")  # Use numpy() for plotting
plt.title("Softmax Activation")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## How Weights Are Initialized

### Default Random Initialization
By default, frameworks like TensorFlow or PyTorch initialize weights randomly. Common initialization techniques include:

#### Glorot Uniform (Xavier) Initialization
This method draws weights from a uniform distribution scaled to the number of inputs and outputs.
$$W ‚àº U(-\sqrt{\frac{6}{n_{in} + n_{out}}}, \sqrt{\frac{6}{n_{in} + n_{out}}})$$

#### He Initialization
This method is specifically designed for ReLU activations and scales weights for better convergence.
$$W ‚àº N(0, \sqrt{\frac{2}{n_{in}}})$$



### Custom Initialization
You can manually set the initial weights using custom initializers to control the starting point of the training process. This can be useful in specific scenarios, but it requires careful consideration.
